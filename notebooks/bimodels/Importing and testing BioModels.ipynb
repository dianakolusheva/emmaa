# Importing and testing BioModels

This notebooks imports models from http://www.ebi.ac.uk/biomodels/ as PySB models and tests them with EMMAA tests.

```python
from pysb.importers.sbml import model_from_biomodels
from pysb.core import ComponentSet
from indra.explanation.model_checker import ModelChecker
from emmaa.model_tests import load_tests_from_s3
from pysb.export.kappa import KappaExporter
```

```python
tests = load_tests_from_s3('small_corpus_tests.pkl')
```

### Examples of different importing errors

```python
model = model_from_biomodels('427')
```

```python
model = model_from_biomodels('20')
```

```python
model = model_from_biomodels('24')
```

### Successful import

These models get imported successfully, but have a problem with tests. First problems are compartments and fixed initials, but even after eiminating those, we still get KappaError.

```python
model = model_from_biomodels('452')
```

```python
model
```

```python
mc = ModelChecker(model)
mc.add_statements([test.stmt for test in tests])
mc.check_model()
```

Compartments are not supported. Try reset the compartments.

```python
model.compartments = ComponentSet([])
```

```python
model
```

```python
mc = ModelChecker(model)
mc.add_statements([test.stmt for test in tests])
mc.check_model()
```

```python
exp = KappaExporter(model)
with open('kappa_452.ka', 'w') as kappa_file:
    kappa_file.write(exp.export())
```

Another example with a different problem

```python
model = model_from_biomodels('659')
```

```python
model
```

```python
model.compartments = ComponentSet([])
mc = ModelChecker(model)
mc.add_statements([test.stmt for test in tests])
mc.check_model()
```

Having fixed initials is a problem too. We can try reset them to False.

```python
model.initials
```

```python
for i in model.initials:
    if i.fixed:
        i.fixed = False
```

```python
model.initials
```

```python
mc = ModelChecker(model)
mc.add_statements([test.stmt for test in tests])
mc.check_model()
```

```python
exp = KappaExporter(model)
with open('kappa_659.ka', 'w') as kappa_file:
    kappa_file.write(exp.export())
```
